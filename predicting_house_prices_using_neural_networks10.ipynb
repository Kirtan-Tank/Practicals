```python
df = df.drop('id', axis = 1)
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)

df= df.drop('date', axis =1)
```

```python
df['zipcode'].value_counts()
```

### Dropping zipcode is too many to call get dummies on hence dropping it

```python
df = df.drop('zipcode',axis =1)
```

```python
sns.heatmap(df.corr())
```

### Making a custom function for identifying renovated or not renovated houses based on the year renovated

```python
df['is_renovated'] = df['yr_renovated'].apply(lambda x: 1 if x != 0 else 0)
```

```python
df= df.drop('yr_renovated', axis =1)
```

### We now start creating a model to predic the price of houses

```python
X = df.drop('price', axis =1).values
y = df.price.values
```

```python
## Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
## scaling the data
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer = 'adam',loss = 'mse')
```

```python
model.fit(x=X_train, y= y_train, validation_data = (X_test,y_test), batch_size = 128 ,epochs= 300);
```

```python
losses = pd.DataFrame(model.history.history)
losses.head()
```

```python
sns.lineplot(x= losses.index, y = losses['loss'])
```

```python
losses.plot()
```

### if we see the lines diverge i.e. the val_loss line start to go up while the loss line keeps going down it would indicate that we are over-fitting the model. This does not happen here. We use early stopping mechanisms to avoid the same.

### We will now evaluate our model

```python
from sklearn.metrics import mean_squared_error, explained_variance_score, mean_absolute_percentage_error
```

```python
predictions = model.predict(X_test)
```

```python
print('mean_squared_error ' ,np.sqrt(mean_squared_error(y_test,predictions)))

print('mean_absolute_percentage_error ',mean_absolute_percentage_error(y_test,predictions))
print(f"Accuracy: {1-mean_absolute_percentage_error(y_test,predictions)}")
```
